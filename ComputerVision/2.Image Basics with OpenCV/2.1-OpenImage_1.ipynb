## Resize

```python
new_img =  cv2.resize(fix_img, (1000,400)) #width-height
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fix_img, (0,0),fix_img,w_ratio ,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

## Flipping

```python
new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```

## Save file

```python
cv2.imwrite('total_new.jpg',fix_img)
```
