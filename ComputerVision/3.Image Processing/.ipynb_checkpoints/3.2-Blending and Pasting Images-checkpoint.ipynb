# Blending and Pasting Images

Let's explore what is commonly known as Arithmetic Image Operations with OpenCV. These are referred to as Arithmetic Operations because OpenCV is simply performing some common math with the pixels for the final effect.

```python
import cv2
```

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('../DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

## BLENDING IMAGES OF THE SAME SIZE
We will blend the values together with the formula:

$$  img1 * \alpha  + img2 * \beta  + \gamma $$

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.8, src2=img2,beta = 0.1,gamma=10)
```

```python
plt.imshow(blended)
```

## OVERLAY SMALL IMAGE ON TOP OF A LARGER IMAGE (NO BLENDING)
### Numpy reassignment

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('../DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0 #position
y_offset = 0 
```

```python
x_end = x_offset + small_img.shape[1] #position
y_end = y_offset + small_img.shape[0] 
```

```python
large_img[y_offset:y_end, x_offset: x_end]  = small_img
```

```python
plt.imshow(large_img)
```
